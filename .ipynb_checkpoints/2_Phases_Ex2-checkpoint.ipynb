```python
def optimize_phase2(prob, x_feasible, u_star, 
                    max_iter=1000, 
                    mu=0.1, 
                    expo_alpha=0.25,
                    expo_lambda=0.75,
                    init_params=1.0, 
                    verbose=True):
    """
    Gii bi ton ESP ( chuyn v BO).
    
    Bi ton:
        Min S(F(x))  [tng ng vi vic ti thiu khong cch ti u_star]
        s.t. x in argmin ||F(x) - P_Q+(F(x))||^2  [x tha mn F(x) in Q+]
             x in C
    """
    
    # Khi to
    x_curr = np.array(x_feasible).copy()
    path_x = [x_curr.copy()]
    
    # Ly cc hm chiu t prob
    P_C = prob.proj_C
    P_Qplus = prob.proj_Qplus
    
    # Cu hnh bng hin th
    if verbose:
        table = PrettyTable()
        table.field_names = ["k", "Alpha_k", "Step_len", "x_curr", "S(F(x))", "Gap Q+"]
        table.float_format = ".4"
        table.align = "r" 


    for k in range(max_iter):
        step = k + 1
        
        fx_curr = prob.objective_func(x_curr) # F(x)
        J = prob.jacobian(x_curr)             # J(x) = F(x)

        # 2. Tnh Gradient Cp thp (v^k) -> Gii quyt rng buc F(x) in Q+
        # r^k = F(x) - P_Q+(F(x))
        proj_q = P_Qplus(fx_curr)
        gap_vector = fx_curr - proj_q
        
        # v^k = J^T * (F(x) - P_Q+(F(x)))
        v_k = J.T @ gap_vector
        
        # 3. Tnh Gradient Cp cao (w^k) -> Ti u ha hm S
        grad_S = fx_curr - u_star
        
        # w^k = J^T *  S(F(x))
        w_k = J.T @ grad_S
        
        # 4. Cp nht tham s
        alpha_k = init_params / (step ** expo_alpha)
        lambda_k = init_params / (step ** expo_lambda)
        
        # 5. Tng hp hng di chuyn d^k
        d_k = v_k + alpha_k * w_k
        
        # 6. Tnh h s chun ha eta_k
        norm_d = np.linalg.norm(d_k)
        eta_k = max(mu, norm_d)
        
        # 7. Bc cp nht v Chiu ln C
        step_size = lambda_k / eta_k
        x_temp = x_curr - step_size * d_k
        x_next = P_C(x_temp)
        
        # --- Lu v Log ---
        val_S = np.linalg.norm(fx_curr - u_star) 
        viol_q = np.linalg.norm(gap_vector)      
        
        if verbose:
            if k % 50 == 0 or k == max_iter - 1:
                x_str = np.array2string(x_curr, precision=3, separator=',')
                table.add_row([k, f"{alpha_k:.4f}", f"{step_size:.4f}", x_str, val_S, viol_q])
        
        path_x.append(x_next.copy())
        x_curr = x_next
        
        # iu kin dng sm (nu cn)
        if norm_d < 1e-6 and viol_q < 1e-6:
            print(f"-> Thut ton hi t sm. Sau {k} vng lp")
            break

    if verbose:
        print(table)
        final_fx = prob.objective_func(x_curr)
        final_viol_q = np.linalg.norm(final_fx - P_Qplus(final_fx))
        print(f"-> Kt qu cui cng Phase 2: {x_curr}")
        print(f"-> S(F(x)): {np.linalg.norm(final_fx - u_star):.4f}")
        print(f"-> Sai s rng buc Q+ (Gap): {final_viol_q:.6f}")

    return x_curr, path_x
```

```python
def visualize_trajectory_(path_x, prob, u_star, 
                         radius_c=2.0, 
                         q_bound=-1.0, 
                         x_limits=(-3.5, 2.5), 
                         y_limits=(-3.5, 1.5),
                         titles=None):
    """
    Hm v qu o ti u ha trong khng gian ngun (C) v khng gian nh (Q+).
    
    Tham s:
    - path_x: List hoc array cha lch s cc im x qua tng bc lp.
    - prob: i tng bi ton (cha phng thc objective_func).
    - u_star: im mc tiu mong mun (trong khng gian nh).
    - radius_c: Bn knh ca tp rng buc C (mc nh 2.0).
    - q_bound: Gii hn trn ca tp Q+ (mc nh -1.0 cho c 2 chiu).
    - x_limits: Tuple (min, max) cho trc ca  th X.
    - y_limits: Tuple (min, max) cho trc ca  th Y.
    - titles: Tuple hoc List cha 2 chui k t (Title_X, Title_Y). 
              V d: ("Khng gian bin", "Khng gian nh").
              Nu None, s dng title mc nh.
    """
    
    # 1. Chun b d liu qu o
    path_x_arr = np.array(path_x) 
    
    # Tnh ton qu o f(x) tng ng
    path_f_arr = np.array([prob.objective_func(p) for p in path_x_arr])

    # 2. TNH TON BIN CA f(C)
    # To cc im trn bin ca C (Hnh trn)  nh x sang bn phi
    theta = np.linspace(0, 2*np.pi, 200) # 200 im cho mt
    boundary_x = radius_c * np.cos(theta)
    boundary_y = radius_c * np.sin(theta)
    boundary_C_points = np.vstack((boundary_x, boundary_y)).T

    # nh x cc im bin ny qua hm f  c bin ca f(C)
    boundary_fC_points = np.array([prob.objective_func(p) for p in boundary_C_points])

    # 3. Khi to khung hnh
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

    # --- X L TIU  (TITLES) ---
    title_x = "Khng gian bin $X$" # Mc nh
    title_y = "Khng gian nh $Y$"  # Mc nh
    
    if titles is not None:
        if len(titles) >= 1: title_x = titles[0]
        if len(titles) >= 2: title_y = titles[1]

    # =========================================================
    # HNH 1: KHNG GIAN BIN X (Tp C v qu o x)
    # =========================================================
    ax1.set_title(title_x, fontsize=14) # <--- S DNG TITLE X

    # V Tp C (Hnh trn)
    circle = plt.Circle((0, 0), radius_c, color='skyblue', alpha=0.3, label='Min C')
    ax1.add_patch(circle)
    ax1.add_patch(plt.Circle((0, 0), radius_c, color='blue', fill=False, linestyle='--')) # Vin

    # V qu o x
    ax1.plot(path_x_arr[:, 0], path_x_arr[:, 1], 'k.-', linewidth=1.5, label='Qu o $x$')
    ax1.scatter(path_x_arr[0, 0], path_x_arr[0, 1], c='cyan', s=100, edgecolors='k', label='Start')
    ax1.scatter(path_x_arr[-1, 0], path_x_arr[-1, 1], c='red', s=120, marker='X', edgecolors='k', label='Final $x$')

    # Trang tr Ax1
    ax1.set_xlim(x_limits[0], x_limits[1])
    ax1.set_ylim(x_limits[0], x_limits[1]) 
    ax1.set_aspect('equal')
    ax1.grid(True, linestyle=':', alpha=0.6)
    ax1.set_xlabel("$x_1$")
    ax1.set_ylabel("$x_2$")
    ax1.legend(loc='upper right')

    # =========================================================
    # HNH 2: KHNG GIAN NH Y (nh f(C), Q+ v qu o f(x))
    # =========================================================
    ax2.set_title(title_y, fontsize=14) # <--- S DNG TITLE Y

    # V Tp nh f(C) dng Polygon
    poly_fC = plt.Polygon(boundary_fC_points, closed=True, facecolor='mediumpurple', alpha=0.4, label='nh $f(C)$')
    ax2.add_patch(poly_fC)
    # V ng vin ca f(C)
    ax2.plot(boundary_fC_points[:, 0], boundary_fC_points[:, 1], color='indigo', linestyle='--', linewidth=1)

    # V bin gii hn ca Q+
    ax2.axvline(x=q_bound, color='red', linestyle='--', linewidth=1.5, label=f'Bin $Q^+$')
    ax2.axhline(y=q_bound, color='red', linestyle='--', linewidth=1.5)

    # T mu vng Q+ (Vng kh thi ca f(x))
    rect_width = abs(y_limits[0] - q_bound) + 5 
    rect_height = abs(y_limits[0] - q_bound) + 5
    rect_Qplus = patches.Rectangle((q_bound - rect_width, q_bound - rect_height), 
                                   rect_width, rect_height, 
                                   color='salmon', alpha=0.1, label='Min $Q^+$')
    ax2.add_patch(rect_Qplus)

    # V im mc tiu u* (Target)
    ax2.scatter(u_star[0], u_star[1], c='gold', s=300, marker='*', edgecolors='k', zorder=10, label='Target $u^*$')

    # V qu o f(x)
    ax2.plot(path_f_arr[:, 0], path_f_arr[:, 1], 'k.-', linewidth=1.5, label='Qu o $f(x)$')
    ax2.scatter(path_f_arr[-1, 0], path_f_arr[-1, 1], c='red', s=120, marker='X', edgecolors='k', label='Final $f(x)$')

    # Trang tr Ax2
    ax2.set_xlim(y_limits[0], y_limits[1])
    ax2.set_ylim(y_limits[0], y_limits[1])
    ax2.set_aspect('equal')
    ax2.grid(True, linestyle=':', alpha=0.6)
    ax2.set_xlabel("$f_1(x)$")
    ax2.set_ylabel("$f_2(x)$")
    ax2.legend(loc='upper right')

    plt.tight_layout()
    plt.show()
```

# Define

```python
# ==========================================
# 1. NH NGHA BI TON C TH
# ==========================================

# Hm f(x) 
def f(x):
    return np.array([x[0]**2, x[1]**2])

# --- Tp C: ---
def c1(x):
    return 5 - (x[0]**2 + x[1]**2)

cons_C = ({'type': 'ineq', 'fun': c1},)
bounds_x = None 
dim_x = 2

# --- Tp Q ---
def q1(y): return np.array([y[0] - 3, y[1] - 1])

def q_plus_upper(t):
    return np.array([3 - t[0], 1 - t[1]])



# Hm dng cho Projection Qplus
cons_Q = (
    {'type': 'eq', 'fun': lambda y: q1(y)},
)

cons_Qplus = (
    {'type': 'ineq', 'fun': q_plus_upper},
)
dim_y = 2
```

# Phase 1

```python
# ==========================================
# 2. KHI TO I TNG V GII CP 1
# ==========================================

print("=== BT U V D MINH HA ESP 2 CP ===")

# Setup Projections
proj_C_handler = Projection(cons=cons_C, bounds=bounds_x, dim=dim_x, proj_type='euclid')
proj_Q_handler = Projection(cons=cons_Q, bounds=None, dim=dim_y, proj_type='qplus')

# Setup Problem
prob = Problem(
    f=[f], jac_f=[jacobian(f)], C=[c1], Q=[], 
    dim_x=dim_x, dim_y=dim_y,
    proj_C=proj_C_handler.project,
    proj_Qplus=proj_Q_handler.project
)

# --- im khi to Phase 1 ---
x_init = np.array([12.0, 12.0]) # Trong C, nhng vi phm Q+

print("\n--- PHASE 1: TM IM KH THI (CQ ALGORITHM) ---")
# Chy thut ton CQ c sn ca bn
x_feasible, x_hist, f_hist, z_proj_hist = CQ_split_acceptance(
    f=prob.objective_func,
    jac_f=prob.jacobian,
    proj_C=prob.proj_C,
    proj_Qplus=prob.proj_Qplus,
    x0=x_init,
    gamma=0.5, # Learning rate cho CQ
    max_iter=50,
    tol=1e-4
)

print(f"-> im kh thi tm c sau Phase 1: {x_feasible}")
```

```python
_ = visualize_complete_system(
    x_hist, f_hist, z_proj_hist,
    cons_C, cons_Qplus,
    x0=x_init,   
)
```

Tm $x \in C $ sao cho $L(x) = \frac{1}{2} \lVert f(x) - u^* \lVert^2$ t min

* im khi to: $x_t, y_t = f(x_t), J_t = J(x_t)$

* $\nabla f$: $w_k = J^T . (y_t - u^*)$
* $\nabla g$: $v_k$
    * $grad_Q = J^T . (y_t - P_{Q^+}(y_t))$
    * $v_k = grad_Q$

* $d_k = \alpha_k * w_k + v_k$

* $
\boxed{
x_{k+1} = x_k - \frac{\lambda_k}{\eta_k} d_k
}
$


trong :  
* $\alpha_k = \frac{1}{k^{0.25}}$
* $\lambda_k = \frac{1}{k^{0.75}}$
* $\eta_k = max\{0.1, \lVert d_k \lVert\}$

```python
def check_constraints(points, constraints, bounds=None):
    """
    Hm ph tr kim tra constraints ( sa li shapes).
    """
    n_points = points.shape[0]
    is_feasible = np.ones(n_points, dtype=bool)
    
    # 1. Kim tra Bounds
    if bounds is not None:
        if isinstance(bounds, (list, tuple)):
            for i, b in enumerate(bounds):
                if b is not None:
                    if b[0] is not None: is_feasible &= (points[:, i] >= b[0])
                    if b[1] is not None: is_feasible &= (points[:, i] <= b[1])
    
    # 2. Kim tra Constraints
    if constraints:
        for cons in constraints:
            vals = None
            
            # A. Th tnh ton vector (nhanh)
            try:
                # Transpose points  khp vi hm nhn vector ct (nu c)
                res = cons['fun'](points.T) 
                
                # X l chiu ca kt qu tr v
                # Nu kt qu l (M, N) -> chuyn thnh (N, M)
                if res.ndim == 2 and res.shape[1] == n_points:
                    vals = res.T
                # Nu kt qu l (N,) -> gi nguyn
                elif res.ndim == 1 and res.shape[0] == n_points:
                    vals = res
                # Nu kt qu l (M,) -> broadcasting c th li, cn fallback
            except:
                pass # Nu li, chuyn sang cch loop

            # B. Fallback: Loop tng im (chm nhng chc)
            if vals is None:
                vals = np.array([cons['fun'](p) for p in points])

            # C. X l logic a rng buc (Multi-constraints)
            # Nu vals c dng (N, M) (tc l mi im c M rng buc con)
            if vals.ndim > 1:
                if cons['type'] == 'ineq':
                    # Tt c cc thnh phn phi >= 0 => min >= 0
                    vals = np.min(vals, axis=1) # Shape tr thnh (N,)
                elif cons['type'] == 'eq':
                    # Tt c cc thnh phn phi ~ 0 => max(abs) ~ 0
                    vals = np.max(np.abs(vals), axis=1) # Shape tr thnh (N,)

            # D. So snh cp nht mask
            # m bo vals by gi c shape (N,) khp vi is_feasible
            if vals.shape[0] != n_points:
                # Trng hp him gp nu hm constraint tr v shape l
                print(f"Warning: Shape mismatch in constraint check. Expected {n_points}, got {vals.shape}")
                continue

            if cons['type'] == 'ineq':
                is_feasible &= (vals >= -1e-5)
            elif cons['type'] == 'eq':
                # Nu l loop fallback, vals cha ly abs, cn ly abs  y nu cha
                if vals.ndim == 1 and np.any(vals < 0): # Check s b
                     is_feasible &= (np.abs(vals) <= 1e-3)
                else:
                     is_feasible &= (vals <= 1e-3)

    return is_feasible
```

```python
from matplotlib.lines import Line2D
def visualize_trajectory_generic(path_x, f_func, 
                                 cons_C, cons_Qplus,
                                 u_star=None,   # <---  thm li tham s ny
                                 bounds_x=None, 
                                 x_limits=(-4, 4), 
                                 y_limits=(-4, 4),
                                 titles=None):
    
    # 1. Chun b d liu
    path_x_arr = np.array(path_x)
    path_f_arr = np.array([f_func(p) for p in path_x_arr])

    # 2. Khi to khung hnh
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))
    
    title_x = "Khng gian bin $X$ (Min $C$)"
    title_y = "Khng gian nh $Y$ (Min $Q^+$)"
    if titles:
        if len(titles) >= 1: title_x = titles[0]
        if len(titles) >= 2: title_y = titles[1]

    # --- HNH 1: KHNG GIAN BIN X ---
    ax1.set_title(title_x, fontsize=13)
    res = 200
    x1_vals = np.linspace(x_limits[0], x_limits[1], res)
    x2_vals = np.linspace(x_limits[0], x_limits[1], res)
    X1, X2 = np.meshgrid(x1_vals, x2_vals)
    grid_points_X = np.vstack([X1.ravel(), X2.ravel()]).T
    
    mask_C = check_constraints(grid_points_X, cons_C, bounds_x).reshape(X1.shape)
    
    ax1.contourf(X1, X2, mask_C, levels=[0.5, 1.5], colors=['skyblue'], alpha=0.4)
    ax1.contour(X1, X2, mask_C, levels=[0.5], colors='blue', linewidths=1.5)
    
    ax1.plot(path_x_arr[:, 0], path_x_arr[:, 1], 'k.-', linewidth=1.5, label='Qu o $x$')
    ax1.scatter(path_x_arr[0, 0], path_x_arr[0, 1], c='cyan', s=80, edgecolors='k', label='Start')
    ax1.scatter(path_x_arr[-1, 0], path_x_arr[-1, 1], c='red', s=100, marker='X', edgecolors='k', label='End')
    
    ax1.set_xlim(x_limits); ax1.set_ylim(x_limits)
    ax1.grid(True, linestyle=':', alpha=0.5); ax1.legend(loc='best')

    # --- HNH 2: KHNG GIAN NH Y ---
    ax2.set_title(title_y, fontsize=13)

    y1_vals = np.linspace(y_limits[0], y_limits[1], res)
    y2_vals = np.linspace(y_limits[0], y_limits[1], res)
    Y1, Y2 = np.meshgrid(y1_vals, y2_vals)
    grid_points_Y = np.vstack([Y1.ravel(), Y2.ravel()]).T
    
    mask_Q = check_constraints(grid_points_Y, cons_Qplus, None).reshape(Y1.shape)
    
    ax2.contourf(Y1, Y2, mask_Q, levels=[0.5, 1.5], colors=['salmon'], alpha=0.2)
    ax2.contour(Y1, Y2, mask_Q, levels=[0.5], colors='red', linewidths=1.5, linestyles='--')
    
    # V m my f(C)
    points_in_C = grid_points_X[mask_C.ravel()]
    if len(points_in_C) > 1500:
        indices = np.random.choice(len(points_in_C), 1500, replace=False)
        sample_C = points_in_C[indices]
    else:
        sample_C = points_in_C
    if len(sample_C) > 0:
        sample_fC = np.array([f_func(p) for p in sample_C])
        ax2.scatter(sample_fC[:, 0], sample_fC[:, 1], c='mediumpurple', s=5, alpha=0.3)

    # V IM MC TIU U_STAR (NU C)
    if u_star is not None:
        ax2.scatter(u_star[0], u_star[1], c='gold', s=200, marker='*', edgecolors='k', zorder=10, label='Target $u^*$')

    ax2.plot(path_f_arr[:, 0], path_f_arr[:, 1], 'k.-', linewidth=1.5, label='Qu o $f(x)$')
    ax2.scatter(path_f_arr[-1, 0], path_f_arr[-1, 1], c='red', s=100, marker='X', edgecolors='k', label='End')

    # Custom Legend
    custom_lines = [Line2D([0], [0], color='blue', lw=2, label='Bin $C$'),
                    Line2D([0], [0], color='red', linestyle='--', lw=2, label='Bin $Q^+$'),
                    Line2D([0], [0], marker='o', color='w', markerfacecolor='mediumpurple', label='Vng $f(C)$', markersize=10)]
    if u_star is not None:
        custom_lines.append(Line2D([0], [0], marker='*', color='w', markerfacecolor='gold', markeredgecolor='k', label='Target $u^*$', markersize=15))

    handles, labels = ax2.get_legend_handles_labels()
    # Lc bt legend trng lp
    ax2.legend(handles + custom_lines, labels + [l.get_label() for l in custom_lines], loc='best')

    ax2.set_xlim(y_limits); ax2.set_ylim(y_limits)
    ax2.grid(True, linestyle=':', alpha=0.5)

    plt.tight_layout()
    plt.show()
```
