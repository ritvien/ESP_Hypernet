# Define

```python
def f(x):
    return np.array([
        4*x[0]**2 + 4*x[1]**2,
        (x[0] - 5)**2 + (x[1] - 5)**2
    ])


#--------------- C --------------------#
def c1(x):
    return -(x[0] - 5)**2 - x[1]**2 + 25
def c2(x):
    return -(x[0] - 8)**2 - (x[1] + 3)**2 + 17.7


#--------------- Q --------------------#
def q1(y):
    return 50**2 - (y[0] - 50)**2 - (y[1] - 50)**2

def q_plus(y):
    center = 50.0
    radius_sq = 50**2  
    
    dx = np.maximum(0, y[0] - center)
    dy = np.maximum(0, y[1] - center)
    return radius_sq - (dx**2 + dy**2)

# Hm dng cho Projection 
cons_C = (
    {
        'type': 'ineq', 'fun' : c1,     
    },
    {
        'type': 'ineq', 'fun' : c2,     
    }
)

dim_x = 2
bounds_x = Bounds([-15,-15],[30, 30])

cons_Q = (
    {
        'type': 'ineq', 'fun' : q1,     
    },    
)

cons_Qplus = (
    {'type': 'ineq', 'fun': q_plus},
)
dim_y = 2
```

```python
x_init = np.array([-10.0, -10.0]) 
gamma = 0.001
max_iter_Phase1 = 50
tol_Phase1 = 1e-4

expo_alpha = 0.35
expo_lambda = 0.75
init_params = 0.5
mu = 0.1
max_iter_Phase2 = 200
```

# Phase 1

```python
# Setup Projections
proj_C_handler = Projection(cons=cons_C, bounds=bounds_x, dim=dim_x, proj_type='euclid')
proj_Q_handler = Projection(cons=cons_Q, bounds=None, dim=dim_y, proj_type='qplus')

# Setup Problem
prob = Problem(
    f=[f], jac_f=[jacobian(f)], C=[], Q=[], 
    dim_x=dim_x, dim_y=dim_y,
    proj_C=proj_C_handler.project,
    proj_Qplus=proj_Q_handler.project
)

print("\n--- PHASE 1: TM IM KH THI (CQ ALGORITHM) ---")
x_feasible, x_hist, f_hist, z_proj_hist = solve_CQ_feasible(
    f=prob.objective_func,
    jac_f=prob.jacobian,
    proj_C=prob.proj_C,
    proj_Qplus=prob.proj_Qplus,
    x0=x_init,
    gamma=gamma, 
    max_iter=max_iter_Phase1,
    tol=tol_Phase1
)

print(f"-> im kh thi tm c sau Phase 1: {x_feasible}")
```

```python
x_final, path_x = optim_Scalarization(
    prob=prob,
    x_feasible=x_feasible,  
    r=test_ray[0], 
    max_iter=max_iter_Phase2,
    mu=mu,
    expo_alpha=expo_alpha,
    expo_lambda=expo_lambda,
    init_params=init_params
)
x_final
```
