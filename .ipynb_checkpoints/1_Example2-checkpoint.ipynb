# Problem

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ==========================================
# 1. NH NGHA D LIU V HM
# ==========================================
d_vec = np.array([0.0, 1.8, 3.0])

def f_map(x1, x2):
    # Tr v ta  y1, y2, y3
    return x1, x2, 2 * x1**2 + 2 * x2**2 + 1

def check_in_Q(y1, y2, y3):
    # Kim tra iu kin ||y - d|| <= 1
    dist_sq = (y1 - d_vec[0])**2 + (y2 - d_vec[1])**2 + (y3 - d_vec[2])**2
    return dist_sq <= 1.0**2

# ==========================================
# 2. TO LI IM  V
# ==========================================

# --- A. To li cho min C (H ta  cc cho hnh trn) ---
r = np.linspace(0, 1, 200)       # Bn knh t 0 n 1
theta = np.linspace(0, 2*np.pi, 60) # Gc quay
R, T = np.meshgrid(r, theta)

# Chuyn sang ta  Descartes (Input x)
X1 = R * np.cos(T)
X2 = R * np.sin(T)

# --- B. Tnh nh f(C) ---
Y1_img, Y2_img, Y3_img = f_map(X1, X2)

# --- C. To li cho qu cu Q ---
u = np.linspace(0, 2 * np.pi, 200)
v = np.linspace(0, np.pi, 200)
# Cng thc tham s ha mt cu
Qx = d_vec[0] + 1.0 * np.outer(np.cos(u), np.sin(v))
Qy = d_vec[1] + 1.0 * np.outer(np.sin(u), np.sin(v))
Qz = d_vec[2] + 1.0 * np.outer(np.ones(np.size(u)), np.cos(v))

# --- D. Tm im giao thoa (Intersection) ---
# Lm phng mng  duyt tng im
flat_y1 = Y1_img.flatten()
flat_y2 = Y2_img.flatten()
flat_y3 = Y3_img.flatten()

# Lc cc im thuc f(C) m nm trong Q
mask = []
for i in range(len(flat_y1)):
    mask.append(check_in_Q(flat_y1[i], flat_y2[i], flat_y3[i]))
mask = np.array(mask)

# Ta  cc im giao nhau
intersect_y1 = flat_y1[mask]
intersect_y2 = flat_y2[mask]
intersect_y3 = flat_y3[mask]

# Tm cc im x tng ng sinh ra giao im ny ( v trn hnh 2D)
flat_x1 = X1.flatten()
flat_x2 = X2.flatten()
intersect_x1 = flat_x1[mask]
intersect_x2 = flat_x2[mask]

# ==========================================
# 3. V  TH
# ==========================================
fig = plt.figure(figsize=(16, 7))

# --- PLOT 1: Min C trong 2D ---
ax1 = fig.add_subplot(121)
ax1.set_title("Min C trong khng gian ngun ($x_1, x_2$)\n$\|x\| \leq 1$", fontsize=12, fontweight='bold')

# V hnh trn n v
circle = plt.Circle((0, 0), 1, color='#A0C4FF', alpha=0.5, label='Min C')
ax1.add_patch(circle)
ax1.plot(intersect_x1, intersect_x2, '.', color='purple', markersize=3, label='Vng nghim ($x$ tha $f(x) \in Q$)')

ax1.set_xlim(-1.5, 1.5)
ax1.set_ylim(-1.5, 1.5)
ax1.set_xlabel("$x_1$")
ax1.set_ylabel("$x_2$")
ax1.set_aspect('equal')
ax1.grid(linestyle='--')
ax1.legend()


# --- PLOT 2: Khng gian ch 3D ---
ax2 = fig.add_subplot(122, projection='3d')
ax2.set_title("nh $f(C)$ v Min $Q$ trong khng gian ch ($y_1, y_2, y_3$)", fontsize=12, fontweight='bold')

# 1. V mt cong f(C)
# Dng plot_surface cho paraboloid
surf = ax2.plot_surface(Y1_img, Y2_img, Y3_img, color='#A0C4FF', alpha=0.4, edgecolor='none')
# Hack  hin th legend cho surface (to proxy artist)
proxy_surf = plt.Rectangle((0, 0), 1, 1, fc="#A0C4FF", alpha=0.4)

# 2. V qu cu Q
# Dng wireframe  nhn xuyn thu c bn trong
ax2.plot_wireframe(Qx, Qy, Qz, color='red', alpha=0.3, linewidth=0.5, label='Bin tp Q (Cu)')

# 3. V im giao nhau
ax2.scatter(intersect_y1, intersect_y2, intersect_y3, color='purple', s=10, depthshade=False, label='Giao $f(C) \cap Q$')

# Trang tr
ax2.set_xlabel('$y_1 (x_1)$')
ax2.set_ylabel('$y_2 (x_2)$')
ax2.set_zlabel('$y_3 (2r^2 + 1)$')

# Thit lp gc nhn  thy r giao im
ax2.view_init(elev=20, azim=-45)

# Legend th cng
ax2.legend(handles=[proxy_surf, 
                    plt.Line2D([0], [0], color='red', lw=0.5, label='Bin tp Q'),
                    plt.Line2D([0], [0], marker='o', color='purple', linestyle='', label='Giao im')],
           loc='upper left')

plt.tight_layout()
plt.show()
```

# Implement

```python
from scipy.optimize import Bounds
from autograd import grad
import autograd.numpy as np  
from autograd import jacobian
from project import Projection
from algorithm_CQ import CQ_split_acceptance
from problem import Problem
```

## Config

```python
def f(x):
    return np.array([
        x[0], 
        x[1], 
        2 * x[0]**2 + 2 * x[1]**2 + 1
    ])

# 2. MIN C (Hnh trn n v)
def c1(x):
    return np.linalg.norm(x) - 1

# 3. MIN Q
d_vec = np.array([0.0, 1.8, 3.0])
def q1(y):
    limit = np.sum(d_vec) + 1
    return np.sum(y) - limit

def q1_dx(y):
    return np.ones_like(y) 
```

```python
cons_C = (
    {
        'type': 'ineq',
        'fun' : lambda x: np.array([-c1(x)]),     
#         'jac' : lambda x: np.array([-c1_dx(x)])   
    },
)


bounds_x = None
dim_x = 2
```

```python
cons_Q = (
    {
        'type': 'ineq',
        'fun' : lambda x: np.array([-q1(x)]),     
#         'jac' : lambda x: np.array([-q1_dx(x)])   
    },
)

bounds_f = None
dim_y = 3
```

```python
cons_Qplus = (
    {
        'type': 'ineq',
        'fun' : lambda x: np.array([q_plus_val(x)]),
        'jac' : lambda x: np.array([q_plus_jac(x)])
    },
)
```

```python
S_limit = np.sum(d_vec) + 1  

def constr_C_func(x):
    return 1.0 - (x[0]**2 + x[1]**2)

def constr_Q_func(y):
    return S_limit - np.sum(y)


cons_C = [
    {'type': 'ineq', 'fun': constr_C_func}
]

cons_Q = [
    {'type': 'ineq', 'fun': constr_Q_func}
]
```

```python
proj_C = Projection(cons=cons_C, bounds=bounds_x, dim=dim_x, proj_type='euclid')

proj_Qplus = Projection(cons=cons_Q, bounds=bounds_f, dim=dim_y, proj_type='qplus')

prob = Problem(
    f=[f],
    jac_f=[jacobian(f)],
    C=[c1],
    Q=[q1],
    dim_x=dim_x,
    dim_y=dim_y,
    proj_C=proj_C.project,
    proj_Qplus=proj_Qplus.project
    
)
```

```python
x0 = [[4.0, 4.0]]
```

## Run

```python
x_opt, x_hist, f_hist, z_proj_hist = CQ_split_acceptance(
                                f=prob.objective_func,
                                jac_f=prob.jacobian,
                                proj_C=prob.proj_C,
                                proj_Qplus=prob.proj_Qplus,
                                x0=x0,
                                gamma=0.05,
                                max_iter=100,
                                tol=1e-10
                            )
```

# Viz

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def visualize_trajectory_on_CQ(x_hist, f_hist, z_proj_hist):
    """
    V chng qu o thut ton ln  th min C v Q.
    
    Input:
    - x_hist: Lch s nghim x (Numpy array Nx2)
    - f_hist: Lch s nh f(x) (Numpy array Nx3)
    - z_proj_hist: Lch s hnh chiu trn Q (Numpy array Nx3)
    """
    
    # Chuyn i input sang numpy array nu cha phi
    x_hist = np.array(x_hist)
    f_hist = np.array(f_hist)
    z_proj_hist = np.array(z_proj_hist)

    # ==========================================
    # 1. TI TO NN (Code gc ca bn)
    # ==========================================
    d_vec = np.array([0.0, 1.8, 3.0])

    def f_map(x1, x2):
        return x1, x2, 2 * x1**2 + 2 * x2**2 + 1

    def check_in_Q(y1, y2, y3):
        # Lu : y l kim tra Q dng CU (theo code bn gi)
        dist_sq = (y1 - d_vec[0])**2 + (y2 - d_vec[1])**2 + (y3 - d_vec[2])**2
        return dist_sq <= 1.0**2

    # --- To li ---
    r = np.linspace(0, 1, 100)       
    theta = np.linspace(0, 2*np.pi, 60) 
    R, T = np.meshgrid(r, theta)
    X1 = R * np.cos(T)
    X2 = R * np.sin(T)
    Y1_img, Y2_img, Y3_img = f_map(X1, X2)

    # --- To li Q (Cu) ---
    u = np.linspace(0, 2 * np.pi, 50) # Gim li cht cho nh
    v = np.linspace(0, np.pi, 50)
    Qx = d_vec[0] + 1.0 * np.outer(np.cos(u), np.sin(v))
    Qy = d_vec[1] + 1.0 * np.outer(np.sin(u), np.sin(v))
    Qz = d_vec[2] + 1.0 * np.outer(np.ones(np.size(u)), np.cos(v))

    # --- Giao im ---
    flat_y1, flat_y2, flat_y3 = Y1_img.flatten(), Y2_img.flatten(), Y3_img.flatten()
    mask = [(check_in_Q(y1, y2, y3)) for y1, y2, y3 in zip(flat_y1, flat_y2, flat_y3)]
    mask = np.array(mask)
    intersect_y1, intersect_y2, intersect_y3 = flat_y1[mask], flat_y2[mask], flat_y3[mask]
    
    flat_x1, flat_x2 = X1.flatten(), X2.flatten()
    intersect_x1, intersect_x2 = flat_x1[mask], flat_x2[mask]

    # ==========================================
    # 2. V  TH
    # ==========================================
    fig = plt.figure(figsize=(16, 8))

    # --- PLOT 1: 2D (Min C + Qu o x) ---
    ax1 = fig.add_subplot(121)
    ax1.set_title("Qu o nghim trn min C ($x_1, x_2$)", fontsize=13, fontweight='bold')

    # Nn tnh
    circle = plt.Circle((0, 0), 1, color='#A0C4FF', alpha=0.4, label='Min C')
    ax1.add_patch(circle)
    ax1.plot(intersect_x1, intersect_x2, '.', color='purple', markersize=2, alpha=0.3, label='Vng kh thi (Feasible)')

    # >>> V QU O X <<<
    # V ng ni
    ax1.plot(x_hist[:, 0], x_hist[:, 1], 'k-', linewidth=1.5, alpha=0.7, label='ng i thut ton')
    # V cc im (mu thay i theo thi gian)
    sc1 = ax1.scatter(x_hist[:, 0], x_hist[:, 1], c=np.arange(len(x_hist)), cmap='autumn', s=40, edgecolors='black', zorder=5)
    # im u v cui
    ax1.scatter(x_hist[0, 0], x_hist[0, 1], marker='*', s=150, color='green', zorder=6, label='Start')
    ax1.scatter(x_hist[-1, 0], x_hist[-1, 1], marker='X', s=150, color='red', zorder=6, label='End')

    ax1.set_xlim(-1.5, 1.5)
    ax1.set_ylim(-1.5, 1.5)
    ax1.set_xlabel("$x_1$")
    ax1.set_ylabel("$x_2$")
    ax1.set_aspect('equal')
    ax1.grid(linestyle='--')
    ax1.legend(loc='upper right')


    # --- PLOT 2: 3D (Khng gian ch + Qu o f(x)) ---
    ax2 = fig.add_subplot(122, projection='3d')
    ax2.set_title("Qu o $f(x)$ hi t v $Q$", fontsize=13, fontweight='bold')

    # Nn tnh
    surf = ax2.plot_surface(Y1_img, Y2_img, Y3_img, color='#A0C4FF', alpha=0.2, edgecolor='none')
    ax2.plot_wireframe(Qx, Qy, Qz, color='red', alpha=0.2, linewidth=0.5)
    ax2.scatter(intersect_y1, intersect_y2, intersect_y3, color='purple', s=5, alpha=0.2)

    # >>> V QU O F(X) <<<
    # V ng i ca f(x)
    ax2.plot(f_hist[:, 0], f_hist[:, 1], f_hist[:, 2], 'k-', linewidth=1.5, label='Qu o $f(x_k)$')
    
    # V cc im f(x)
    sc2 = ax2.scatter(f_hist[:, 0], f_hist[:, 1], f_hist[:, 2], c=np.arange(len(f_hist)), cmap='autumn', s=30, depthshade=False)
    
    # V im chiu P_Q(f(x))  thy ch n ti mi bc
    # Ch v vi im cui hoc v m   ri
    if len(z_proj_hist) > 0:
        ax2.scatter(z_proj_hist[:, 0], z_proj_hist[:, 1], z_proj_hist[:, 2], marker='x', color='blue', s=20, label='$P_Q(f(x))$')
        
        # V on thng ni f(x) -> z_proj (th hin sai s khong cch)
        # Ch v 20 im u tin v im cui cng  trnh ri hnh nu iter ln
        indices_to_draw = list(range(min(20, len(f_hist)))) + [len(f_hist)-1]
        for i in indices_to_draw:
            ax2.plot([f_hist[i, 0], z_proj_hist[i, 0]], 
                     [f_hist[i, 1], z_proj_hist[i, 1]], 
                     [f_hist[i, 2], z_proj_hist[i, 2]], 
                     color='blue', linestyle=':', linewidth=0.8)

    ax2.set_xlabel('$y_1$')
    ax2.set_ylabel('$y_2$')
    ax2.set_zlabel('$y_3$')
    ax2.view_init(elev=20, azim=-45)
    
    # To legend gi cho 3D
    proxy_surf = plt.Rectangle((0, 0), 1, 1, fc="#A0C4FF", alpha=0.4, label='nh $f(C)$')
    proxy_line = plt.Line2D([0], [0], color='k', lw=1.5, label='Qu o')
    proxy_proj = plt.Line2D([0], [0], color='blue', linestyle=':', label='Khong cch n Q')
    
    ax2.legend(handles=[proxy_surf, proxy_line, proxy_proj], loc='upper left')

    plt.tight_layout()
    plt.show()
```

# V li lung thut ton

<font size=5>Tm $x\in C$ sao cho $\exists y \in Q$ tha mn $f(x) \leq y \quad(1)$ 

trong : 
* $C, Q$ l tp li, ng
* $f$ l nh x li

---

nh ngha: 

<font size=4> $Q^+ = \{z \in R^m | \exists y \in Q, z \leq y\}$

Khi :

<font size=4> $ (1) \Leftrightarrow $ Tm $x \in C$ sao cho $f(x) \in Q^+$

---

Hm khong cch:

$ \Phi(x) := \frac{1}{2} \rVert f(x) - P_{Q^+}(f(x)\rVert^2  $

Mc tiu: a $f(x)$ v gn vi hnh chiu ca n trn tp $Q^+$ nht

* $\Phi(x) = 0 \rightarrow x$ l nghim chp nhn tch
* $\Phi(x) \neq 0 \rightarrow $ chuyn v bi ton xp x tt nht

---


<font size=5> $\nabla \Phi (x) = J_f(x)^T (I - P_{Q^+})f(x)$

Cng thc cp nht $x^k$

<font size=5> $\Rightarrow x^{k+1} = P_C(x^k - \gamma_k \nabla\Phi(x^k))$ 

---

$P_C(\tilde x) = \text{Argmin}_{x \in C} \rVert x - \tilde x \rVert ^2$

---

$y^* = \text{Argmin}_{y \in Q} \rVert (y - \tilde z )_+\rVert ^2$

$P_{Q^+}(\tilde z) = \text{min}(\tilde z, y^*) = \tilde z - (\tilde z - y^*)_+$
