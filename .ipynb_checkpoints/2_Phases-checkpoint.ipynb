```python
import autograd.numpy as np
from autograd import jacobian
from scipy.optimize import Bounds
from project import Projection
from algorithm_CQ import CQ_split_acceptance
from problem import Problem
from prettytable import PrettyTable
from utils import visualize_complete_system
from utils import visualize_trajectory

import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
def optimize_phase2(prob, x_feasible, u_star, 
                    max_iter=1000, 
                    mu=0.1, 
                    expo_alpha=0.25,
                    expo_lambda=0.75,
                    init_params=1.0, 
                    verbose=True):
    
    # Khi to
    x_curr = np.array(x_feasible).copy()
    path_x = [x_curr.copy()]
    
    # Ly cc hm chiu t prob
    P_C = prob.proj_C
    P_Qplus = prob.proj_Qplus
    
    # Cu hnh bng hin th
    if verbose:
        table = PrettyTable()
        table.field_names = ["k", "Params (a(w), b(v))", "x_curr", "dist_u*", "Viol_C", "Viol_Q+"]
        # nh dng ct cho p
        table.float_format = ".4"
        table.align = "r" 

    print(f"\n--- BT U PHASE 2: Di chuyn t {x_curr} v {u_star} ---")

    for k in range(max_iter):
        step = k + 1
        
        obj_curr = prob.objective_func(x_curr) # f(x)
        J = prob.jacobian(x_curr)              # J(x)

        w_k = J.T @ (obj_curr - u_star)
        
        proj_q = P_Qplus(obj_curr)
        grad_Q = J.T @ (obj_curr - proj_q)
        
        # v_C = x - P_C(x)
        proj_c = P_C(x_curr)
        grad_C = (x_curr - proj_c)
        
        v_k = grad_C + grad_Q
        
        alpha_k = init_params / (step ** 0.25)
        lambda_k = init_params / (step ** 0.75)
        
        d_k = alpha_k * w_k + v_k
        
        norm_d = np.linalg.norm(d_k)
        eta_k = max(mu, norm_d)
        
        step_vector = (lambda_k / eta_k) * d_k
        x_next = x_curr - step_vector
        
        # --- Lu v Log ---
        # Tnh ton cc ch s  theo di
        dist_u = np.linalg.norm(obj_curr - u_star)
        viol_c = np.linalg.norm(grad_C)
        viol_q = np.linalg.norm(obj_curr - proj_q) 
        
        if verbose:
            if k % 20 == 0 or k == max_iter - 1:
                x_str = np.array2string(x_curr, precision=3, separator=',')
                table.add_row([k, f"({alpha_k*(lambda_k / eta_k):.3f}, {(lambda_k / eta_k):.3f})", x_str, dist_u, viol_c, viol_q])
        
        path_x.append(x_next.copy())
        x_curr = x_next

    if verbose:
        print(table)
        print(f"-> Kt qu cui cng Phase 2: {x_curr}")
        print(f"-> Khong cch ti u*: {np.linalg.norm(prob.objective_func(x_curr) - u_star):.4f}")

    return x_curr, path_x
```

```python
# ==========================================
# 1. NH NGHA BI TON C TH
# ==========================================

# Hm f(x) = x
def f(x):
    return np.array([x[0], x[1]])

def jac_f_func(x):
    return np.eye(2)

# --- Tp C: Hnh trn tm O, bn knh 2 ---
# x1^2 + x2^2 <= 4
def c1(x):
    return 4 - (x[0]**2 + x[1]**2)

cons_C = ({'type': 'ineq', 'fun': c1},)
bounds_x = None 
dim_x = 2

# --- Tp Q: Hnh hp [-3, -1] x [-3, -1] ---
def q_box_lower(y): return np.array([y[0] + 3, y[1] + 3])
def q_box_upper(y): return np.array([-1 - y[0], -1 - y[1]])
def q_plus_upper(t):
    return np.array([-1 - t[0], -1 - t[1]])



# Hm dng cho Projection Qplus (ch cn chn trn ca Q  xc nh Q+)
# Q+ = {z | z <= y_max},  y y_max = (-1, -1)
cons_Q = (
    {'type': 'ineq', 'fun': lambda y: np.concatenate([q_box_lower(y), q_box_upper(y)])},
)

cons_Qplus = (
    {'type': 'ineq', 'fun': q_plus_upper},
)
dim_y = 2
```

```python
# ==========================================
# 2. KHI TO I TNG V GII CP 1
# ==========================================

print("=== BT U V D MINH HA ESP 2 CP ===")

# Setup Projections
proj_C_handler = Projection(cons=cons_C, bounds=bounds_x, dim=dim_x, proj_type='euclid')
proj_Q_handler = Projection(cons=cons_Q, bounds=None, dim=dim_y, proj_type='qplus')

# Setup Problem
prob = Problem(
    f=[f], jac_f=[jac_f_func], C=[c1], Q=[], 
    dim_x=dim_x, dim_y=dim_y,
    proj_C=proj_C_handler.project,
    proj_Qplus=proj_Q_handler.project
)

# --- im khi to Phase 1 ---
x_init = np.array([2.0, 2.0]) # Trong C, nhng vi phm Q+

print("\n--- PHASE 1: TM IM KH THI (CQ ALGORITHM) ---")
# Chy thut ton CQ c sn ca bn
x_feasible, x_hist, f_hist, z_proj_hist = CQ_split_acceptance(
    f=prob.objective_func,
    jac_f=prob.jacobian,
    proj_C=prob.proj_C,
    proj_Qplus=prob.proj_Qplus,
    x0=x_init,
    gamma=0.5, # Learning rate cho CQ
    max_iter=50,
    tol=1e-4
)

print(f"-> im kh thi tm c sau Phase 1: {x_feasible}")
```

```python
_ = visualize_complete_system(
    x_hist, f_hist, z_proj_hist,
    cons_C, cons_Qplus,
    x0=x_init,   
)
```

* im khi to: $x_t, y_t = f(x_t), J_t = J(x_t)$

* Grad f: $w_k = J^T . (y_t - u^*)$
* Grad g: $v_k$
    * $grad_C = x_t - P_C(x_t)$
    * $grad_Q = J^T . (y_t - P_{Q^+}(y_t))$
    * $v_k = grad_C + grad_Q$

* $d_k = \alpha_k * w_k + v_k$

* $
\boxed{
x_{k+1} = x_k - \frac{\lambda_k}{\eta_k} d_k
}
$


trong :  
* $\alpha_k = \frac{1}{k^{0.25}}$
* $\lambda_k = \frac{1}{k^{0.75}}$
* $\eta_k = max\{0.1, \lVert d_k \lVert\}$

```python
x_final, path_x = optimize_phase2(
    prob=prob,
    x_feasible=x_feasible,  
    u_star=np.array([-0.5, -1.5]), 
    max_iter=1000,
    mu=0.1,
    expo_alpha=0.25,
    expo_lambda=0.75,
    init_params=1.0,
    verbose=True
)
```
