# Implement

```python
import autograd.numpy as np
from autograd import jacobian
from scipy.optimize import Bounds
from project import Projection
from algorithm_CQ import CQ_split_acceptance
from problem import Problem
from prettytable import PrettyTable
from utils import visualize_complete_system
from utils import visualize_trajectory

import matplotlib.pyplot as plt
import matplotlib.patches as patches

from prettytable import PrettyTable
import itertools
import pandas as pd
```

```python
def optimize_phase2(prob, x_feasible, u_star, 
                    max_iter=1000, 
                    mu=0.1, 
                    expo_alpha=0.25,
                    expo_lambda=0.75,
                    init_params=1.0, 
                    verbose=True):
    
    # Khi to
    x_curr = np.array(x_feasible).copy()
    path_x = [x_curr.copy()]
    
    # Ly cc hm chiu t prob
    P_C = prob.proj_C
    P_Qplus = prob.proj_Qplus
    
    # Cu hnh bng hin th
    if verbose:
        table = PrettyTable()
        table.field_names = ["k", "Params (a(w), b(v))", "x_curr", "dist_u*", "Viol_C", "Viol_Q+"]
        # nh dng ct cho p
        table.float_format = ".4"
        table.align = "r" 

    print(f"\n--- BT U PHASE 2: Di chuyn t {x_curr} v {u_star} ---")

    for k in range(max_iter):
        step = k + 1
        
        obj_curr = prob.objective_func(x_curr) # f(x)
        J = prob.jacobian(x_curr)              # J(x)

        w_k = J.T @ (obj_curr - u_star)
        
        proj_q = P_Qplus(obj_curr)
        grad_Q = J.T @ (obj_curr - proj_q)
        
        # v_C = x - P_C(x)
        proj_c = P_C(x_curr)
        grad_C = (x_curr - proj_c)
        
        v_k = grad_Q
        
        alpha_k = init_params / (step ** expo_alpha)
        lambda_k = init_params / (step ** expo_lambda)
        
        d_k = alpha_k * w_k + v_k
        
        norm_d = np.linalg.norm(d_k)
        eta_k = max(mu, norm_d)
        
        step_vector = (lambda_k / eta_k) * d_k
        x_next = x_curr - step_vector
        
        # --- Lu v Log ---
        # Tnh ton cc ch s  theo di
        dist_u = np.linalg.norm(obj_curr - u_star)
        viol_c = np.linalg.norm(grad_C)
        viol_q = np.linalg.norm(obj_curr - proj_q) 
        
        if verbose:
            if k % 20 == 0 or k == max_iter - 1:
                x_str = np.array2string(x_curr, precision=3, separator=',')
                table.add_row([k, f"({alpha_k*(lambda_k / eta_k):.3f}, {(lambda_k / eta_k):.3f})", x_str, dist_u, viol_c, viol_q])
        
        path_x.append(x_next.copy())
        x_curr = x_next

    if verbose:
        print(table)
        print(f"-> Kt qu cui cng Phase 2: {x_curr}")
        print(f"-> Khong cch ti u*: {np.linalg.norm(prob.objective_func(x_curr) - u_star):.4f}")

    return x_curr, path_x
```

```python
def visualize_trajectory_(path_x, prob, u_star, 
                         radius_c=2.0, 
                         q_bound=-1.0, 
                         x_limits=(-3.5, 2.5), 
                         y_limits=(-3.5, 1.5),
                         titles=None):
    """
    Hm v qu o ti u ha trong khng gian ngun (C) v khng gian nh (Q+).
    
    Tham s:
    - path_x: List hoc array cha lch s cc im x qua tng bc lp.
    - prob: i tng bi ton (cha phng thc objective_func).
    - u_star: im mc tiu mong mun (trong khng gian nh).
    - radius_c: Bn knh ca tp rng buc C (mc nh 2.0).
    - q_bound: Gii hn trn ca tp Q+ (mc nh -1.0 cho c 2 chiu).
    - x_limits: Tuple (min, max) cho trc ca  th X.
    - y_limits: Tuple (min, max) cho trc ca  th Y.
    - titles: Tuple hoc List cha 2 chui k t (Title_X, Title_Y). 
              V d: ("Khng gian bin", "Khng gian nh").
              Nu None, s dng title mc nh.
    """
    
    # 1. Chun b d liu qu o
    path_x_arr = np.array(path_x) 
    
    # Tnh ton qu o f(x) tng ng
    path_f_arr = np.array([prob.objective_func(p) for p in path_x_arr])

    # 2. TNH TON BIN CA f(C)
    # To cc im trn bin ca C (Hnh trn)  nh x sang bn phi
    theta = np.linspace(0, 2*np.pi, 200) # 200 im cho mt
    boundary_x = radius_c * np.cos(theta)
    boundary_y = radius_c * np.sin(theta)
    boundary_C_points = np.vstack((boundary_x, boundary_y)).T

    # nh x cc im bin ny qua hm f  c bin ca f(C)
    boundary_fC_points = np.array([prob.objective_func(p) for p in boundary_C_points])

    # 3. Khi to khung hnh
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

    # --- X L TIU  (TITLES) ---
    title_x = "Khng gian bin $X$" # Mc nh
    title_y = "Khng gian nh $Y$"  # Mc nh
    
    if titles is not None:
        if len(titles) >= 1: title_x = titles[0]
        if len(titles) >= 2: title_y = titles[1]

    # =========================================================
    # HNH 1: KHNG GIAN BIN X (Tp C v qu o x)
    # =========================================================
    ax1.set_title(title_x, fontsize=14) # <--- S DNG TITLE X

    # V Tp C (Hnh trn)
    circle = plt.Circle((0, 0), radius_c, color='skyblue', alpha=0.3, label='Min C')
    ax1.add_patch(circle)
    ax1.add_patch(plt.Circle((0, 0), radius_c, color='blue', fill=False, linestyle='--')) # Vin

    # V qu o x
    ax1.plot(path_x_arr[:, 0], path_x_arr[:, 1], 'k.-', linewidth=1.5, label='Qu o $x$')
    ax1.scatter(path_x_arr[0, 0], path_x_arr[0, 1], c='cyan', s=100, edgecolors='k', label='Start')
    ax1.scatter(path_x_arr[-1, 0], path_x_arr[-1, 1], c='red', s=120, marker='X', edgecolors='k', label='Final $x$')

    # Trang tr Ax1
    ax1.set_xlim(x_limits[0], x_limits[1])
    ax1.set_ylim(x_limits[0], x_limits[1]) 
    ax1.set_aspect('equal')
    ax1.grid(True, linestyle=':', alpha=0.6)
    ax1.set_xlabel("$x_1$")
    ax1.set_ylabel("$x_2$")
    ax1.legend(loc='upper right')

    # =========================================================
    # HNH 2: KHNG GIAN NH Y (nh f(C), Q+ v qu o f(x))
    # =========================================================
    ax2.set_title(title_y, fontsize=14) # <--- S DNG TITLE Y

    # V Tp nh f(C) dng Polygon
    poly_fC = plt.Polygon(boundary_fC_points, closed=True, facecolor='mediumpurple', alpha=0.4, label='nh $f(C)$')
    ax2.add_patch(poly_fC)
    # V ng vin ca f(C)
    ax2.plot(boundary_fC_points[:, 0], boundary_fC_points[:, 1], color='indigo', linestyle='--', linewidth=1)

    # V bin gii hn ca Q+
    ax2.axvline(x=q_bound, color='red', linestyle='--', linewidth=1.5, label=f'Bin $Q^+$')
    ax2.axhline(y=q_bound, color='red', linestyle='--', linewidth=1.5)

    # T mu vng Q+ (Vng kh thi ca f(x))
    rect_width = abs(y_limits[0] - q_bound) + 5 
    rect_height = abs(y_limits[0] - q_bound) + 5
    rect_Qplus = patches.Rectangle((q_bound - rect_width, q_bound - rect_height), 
                                   rect_width, rect_height, 
                                   color='salmon', alpha=0.1, label='Min $Q^+$')
    ax2.add_patch(rect_Qplus)

    # V im mc tiu u* (Target)
    ax2.scatter(u_star[0], u_star[1], c='gold', s=300, marker='*', edgecolors='k', zorder=10, label='Target $u^*$')

    # V qu o f(x)
    ax2.plot(path_f_arr[:, 0], path_f_arr[:, 1], 'k.-', linewidth=1.5, label='Qu o $f(x)$')
    ax2.scatter(path_f_arr[-1, 0], path_f_arr[-1, 1], c='red', s=120, marker='X', edgecolors='k', label='Final $f(x)$')

    # Trang tr Ax2
    ax2.set_xlim(y_limits[0], y_limits[1])
    ax2.set_ylim(y_limits[0], y_limits[1])
    ax2.set_aspect('equal')
    ax2.grid(True, linestyle=':', alpha=0.6)
    ax2.set_xlabel("$f_1(x)$")
    ax2.set_ylabel("$f_2(x)$")
    ax2.legend(loc='upper right')

    plt.tight_layout()
    plt.show()
```

# Define

```python
# ==========================================
# 1. NH NGHA BI TON C TH
# ==========================================

# Hm f(x) = x
def f(x):
    return np.array([x[0], x[1]])

def jac_f_func(x):
    return np.eye(2)

# --- Tp C: Hnh trn tm O, bn knh 2 ---
# x1^2 + x2^2 <= 4
def c1(x):
    return 4 - (x[0]**2 + x[1]**2)

cons_C = ({'type': 'ineq', 'fun': c1},)
bounds_x = None 
dim_x = 2

# --- Tp Q: Hnh hp [-3, -1] x [-3, -1] ---
def q_box_lower(y): return np.array([y[0] + 3, y[1] + 3])
def q_box_upper(y): return np.array([-1 - y[0], -1 - y[1]])
def q_plus_upper(t):
    return np.array([-1 - t[0], -1 - t[1]])



# Hm dng cho Projection Qplus (ch cn chn trn ca Q  xc nh Q+)
# Q+ = {z | z <= y_max},  y y_max = (-1, -1)
cons_Q = (
    {'type': 'ineq', 'fun': lambda y: np.concatenate([q_box_lower(y), q_box_upper(y)])},
)

cons_Qplus = (
    {'type': 'ineq', 'fun': q_plus_upper},
)
dim_y = 2
```

# Phase 1

```python
# ==========================================
# 2. KHI TO I TNG V GII CP 1
# ==========================================

print("=== BT U V D MINH HA ESP 2 CP ===")

# Setup Projections
proj_C_handler = Projection(cons=cons_C, bounds=bounds_x, dim=dim_x, proj_type='euclid')
proj_Q_handler = Projection(cons=cons_Q, bounds=None, dim=dim_y, proj_type='qplus')

# Setup Problem
prob = Problem(
    f=[f], jac_f=[jac_f_func], C=[c1], Q=[], 
    dim_x=dim_x, dim_y=dim_y,
    proj_C=proj_C_handler.project,
    proj_Qplus=proj_Q_handler.project
)

# --- im khi to Phase 1 ---
x_init = np.array([2.0, 2.0]) # Trong C, nhng vi phm Q+

print("\n--- PHASE 1: TM IM KH THI (CQ ALGORITHM) ---")
# Chy thut ton CQ c sn ca bn
x_feasible, x_hist, f_hist, z_proj_hist = CQ_split_acceptance(
    f=prob.objective_func,
    jac_f=prob.jacobian,
    proj_C=prob.proj_C,
    proj_Qplus=prob.proj_Qplus,
    x0=x_init,
    gamma=0.5, # Learning rate cho CQ
    max_iter=50,
    tol=1e-4
)

print(f"-> im kh thi tm c sau Phase 1: {x_feasible}")
```

```python
_ = visualize_complete_system(
    x_hist, f_hist, z_proj_hist,
    cons_C, cons_Qplus,
    x0=x_init,   
)
```

* im khi to: $x_t, y_t = f(x_t), J_t = J(x_t)$

* Grad f: $w_k = J^T . (y_t - u^*)$
* Grad g: $v_k$
    * $grad_Q = J^T . (y_t - P_{Q^+}(y_t))$
    * $v_k = grad_Q$

* $d_k = \alpha_k * w_k + v_k$

* $
\boxed{
x_{k+1} = x_k - \frac{\lambda_k}{\eta_k} d_k
}
$


trong :  
* $\alpha_k = \frac{1}{k^{0.25}}$
* $\lambda_k = \frac{1}{k^{0.75}}$
* $\eta_k = max\{0.1, \lVert d_k \lVert\}$

# Run all

```python
def run_experiment_grid(prob, x_feasible, 
                        list_u_star, 
                        list_expo_alpha, 
                        list_expo_lambda, 
                        list_init_params,
                        max_iter=1000):
    
    # To tt c cc t hp tham s (Cartesian product)
    combinations = list(itertools.product(list_u_star, list_expo_alpha, list_expo_lambda, list_init_params))
    
    print(f"=== TNG S TH NGHIM CN CHY: {len(combinations)} ===")
    
    summary_data = []
    
    for i, (u_star, exp_a, exp_l, init_p) in enumerate(combinations):
        exp_id = i + 1
        print(f"\n{'='*60}")
        print(f"EXPERIMENT #{exp_id}/{len(combinations)}")
        print(f"Config: u*={u_star}, exp_a={exp_a}, exp_l={exp_l}, init={init_p}")
        print(f"{'='*60}")
        
        # 1. Chy ti u
        x_final, path_x = optimize_phase2(
            prob=prob,
            x_feasible=x_feasible,
            u_star=u_star,
            max_iter=max_iter,
            mu=0.1,
            init_params=init_p,
            expo_alpha=exp_a,   # Truyn tham s ng
            expo_lambda=exp_l,  # Truyn tham s ng
            verbose=False       # Tt verbose chi tit   ri
        )
        
        # 2. Tnh ton kt qu cui
        final_f = prob.objective_func(x_final)
        dist_u = np.linalg.norm(final_f - u_star)
        
        # Tnh vi phm rng buc cui cng
        proj_q = prob.proj_Qplus(final_f)
        viol_q = np.linalg.norm(final_f - proj_q)
        proj_c = prob.proj_C(x_final)
        viol_c = np.linalg.norm(x_final - proj_c)
        
        # Lu vo bng tng kt
        summary_data.append({
            "ID": exp_id,
            "u*": str(u_star),
            "Exp_Alpha": exp_a,
            "Exp_Lambda": exp_l,
            "Init_P": init_p,
            "Final_Dist_u*": round(dist_u, 5),
            "Viol_Q+": round(viol_q, 5),
            "Viol_C": round(viol_c, 5)
        })
        
        # 3. V  th
        # To title ng cho tng hnh
        title_left = f"Exp #{exp_id}: Params(a={exp_a}, l={exp_l})"
        title_right = f"Target {u_star} | Init {init_p} | Dist: {dist_u:.3f}"
        
        visualize_trajectory_(
            path_x=path_x,      
            prob=prob,           
            u_star=u_star,      
            radius_c=2.0,       
            q_bound=-1.0,       
            x_limits=(-2.5, 2.5), 
            y_limits=(-2.5, 0.5), # Cn chnh li cho hp l vi u*
            titles=(title_left, title_right)
        )
        
    # 4. Hin th bng tng kt sau khi chy xong ht
    print("\n\n" + "="*30 + " KT QU TNG HP " + "="*30)
    df_results = pd.DataFrame(summary_data)
    
    # Hin th p bng PrettyTable nu khng dng Jupyter Notebook, 
    # hoc print dataframe nu dng Jupyter
    if len(summary_data) > 0:
        summary_table = PrettyTable()
        summary_table.field_names = df_results.columns.tolist()
        for row in df_results.itertuples(index=False):
            summary_table.add_row(row)
        print(summary_table)
        
    return df_results
```

```python
# --- CU HNH TH NGHIM ---

# 1. Cc im mc tiu mun th
list_u_star = [
    np.array([-0.5, -1.5]), # im 1
    np.array([-2.0, -0.5])  # im 2
]


list_expo_alpha = [0.25, 0.5]


list_expo_lambda = [0.60, 0.85]

list_init_params = [1.0, 2.0]

# --- CHY ---
df_results = run_experiment_grid(
    prob=prob,
    x_feasible=x_feasible,
    list_u_star=list_u_star,
    list_expo_alpha=list_expo_alpha,
    list_expo_lambda=list_expo_lambda,
    list_init_params=list_init_params,
    max_iter=500 
)
```
