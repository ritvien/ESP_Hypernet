# Problem

```python
def f1(x):
    return 4*x[0]**2 + 4*x[1]**2
def f2(x):
    return (x[0] - 5)**2 + (x[1] - 5)**2

#--------------- C --------------------#
def c1(x):
    return (x[0] - 5)**2 + x[1]**2 - 25
def c2(x):
    return -(x[0] - 8)**2 - (x[1] + 3)**2 + 7.7


#--------------- Q --------------------#
def q1(x):
    return -f1(x) + 20
def q2(x):
    return f1(x) - 150

def q3(x):
    return -f2(x) + 10
def q4(x):
    return f2(x) - 40

```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# 1. nh ngha cc hm
def f1_vec(X, Y):
    return 4*X**2 + 4*Y**2

def f2_vec(X, Y):
    return (X - 5)**2 + (Y - 5)**2

# 2. Chun b d liu cho PLOT 1 (Min C trn khng gian x1, x2)
x_range = np.linspace(-6, 12, 600)
y_range = np.linspace(-8, 12, 600)
X, Y = np.meshgrid(x_range, y_range)

# Logic min C
val_c1 = (X - 5)**2 + Y**2 - 25
val_c2 = -(X - 8)**2 - (Y + 3)**2 + 7.7
mask_C = (val_c1 <= 0) & (val_c2 <= 0)

# 3. Chun b d liu cho PLOT 2 (nh x sang khng gian f1, f2)
# Sinh ngu nhin im  biu din "nh ca min C" ln trc f
num_samples = 30000
x_rand = np.random.uniform(-6, 12, num_samples)
y_rand = np.random.uniform(-8, 12, num_samples)

# Lc ly cc im thuc min C
c1_rand = (x_rand - 5)**2 + y_rand**2 - 25
c2_rand = -(x_rand - 8)**2 - (y_rand + 3)**2 + 7.7
mask_C_rand = (c1_rand <= 0) & (c2_rand <= 0)

x_valid = x_rand[mask_C_rand]
y_valid = y_rand[mask_C_rand]

# Tnh gi tr f1, f2 tng ng
f1_vals = f1_vec(x_valid, y_valid)
f2_vals = f2_vec(x_valid, y_valid)


# 4. V  th
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

# ==========================================
# PLOT 1: MIN C (Gi nguyn: Khng gian bin x)
# ==========================================
ax1.set_title("Min C trong khng gian bin $(x_1, x_2)$", fontsize=14, fontweight='bold')
ax1.contourf(X, Y, mask_C, levels=[0.5, 1.5], colors=['#A0C4FF'], alpha=0.6)
ax1.contour(X, Y, val_c1, levels=[0], colors='blue', linewidths=2)
ax1.contour(X, Y, val_c2, levels=[0], colors='red', linestyles='dashed', linewidths=2)

ax1.set_xlabel("$x_1$", fontsize=12)
ax1.set_ylabel("$x_2$", fontsize=12)
ax1.axhline(0, color='black', linewidth=0.5)
ax1.axvline(0, color='black', linewidth=0.5)
ax1.grid(True, linestyle='--', alpha=0.5)
ax1.set_aspect('equal')

# Legend cho ax1
patch_c = mpatches.Patch(color='#A0C4FF', label='Vng kh thi C')
ax1.legend(handles=[patch_c], loc='upper left')


# ==========================================
# PLOT 2: MIN Q (Sa i: Khng gian mc tiu f)
# ==========================================
ax2.set_title("Min Q trong khng gian mc tiu $(f_1, f_2)$", fontsize=14, fontweight='bold')

# a. V nh ca min C (Scatter plot)
# y l tp hp gi tr (f1, f2) m x trong min C to ra
ax2.scatter(f1_vals, f2_vals, s=5, color='#A0C4FF', alpha=0.5, label='nh ca min C ($f(C)$)')

# b. V min Q (Hnh ch nht)
# Q: 20 <= f1 <= 150, 10 <= f2 <= 40
# Rectangle(xy, width, height)
rect_Q = mpatches.Rectangle((20, 10), 130, 30, 
                            linewidth=2, edgecolor='red', facecolor='none', 
                            linestyle='--', label='Hp Q ($20 \leq f_1 \leq 150, ...$)')
# T mu nht cho hp Q
ax2.add_patch(mpatches.Rectangle((20, 10), 130, 30, color='#FFADAD', alpha=0.3))
ax2.add_patch(rect_Q)

# c. Lm ni bt phn giao nhau (nghim chp nhn c)
mask_intersection = (f1_vals >= 20) & (f1_vals <= 150) & (f2_vals >= 10) & (f2_vals <= 40)
ax2.scatter(f1_vals[mask_intersection], f2_vals[mask_intersection], 
            s=5, color='purple', label='Giao ca C v Q')

# Trang tr ax2
ax2.set_xlabel("$f_1(x) = 4x_1^2 + 4x_2^2$", fontsize=12)
ax2.set_ylabel("$f_2(x) = (x_1-5)^2 + (x_2-5)^2$", fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.5)

# Thit lp gii hn trc sao cho nhn thy c m my im v hp Q
ax2.set_xlim(-10, 450)
ax2.set_ylim(-10, 120)

ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

# Implement

```python
from scipy.optimize import minimize
from scipy.optimize import Bounds
from scipy.optimize import BFGS
from autograd import grad
import numpy as np
from numpy import linalg as LA
from prettytable import PrettyTable
from tqdm import tqdm
```

```python
class Problem():
    def __init__(self, f, jac_f, C, Q, proj_C, proj_Qplus):
        self.f = f
        self.jac_f = jac_f
        self.C = C
        self.Q = Q
        self.proj_C = proj_C
        self.proj_Qplus = proj_Qplus
    
    def objective_func(self, x):
        return np.array([func(x) for func in self.f])
    
    def jacobian(self, x):
        return np.array([func(x) for func in self.jac_f])
            
            
        
class Projection:
    def __init__(self, cons, bounds, n, proj_type='euclid'):
        """
        cons: Rng buc (constraints)
        bounds: Gii hn bin (bounds)
        n: S chiu
        proj_type: Loi php chiu ('euclid' cho min C hoc 'qplus' cho min Q+)
        """
        self.cons = cons
        self.bounds = bounds
        self.n = n
        self.proj_type = proj_type
        
        # Cu hnh hm mc tiu da trn loi php chiu
        if self.proj_type == 'qplus':
            self.objective_func = self._obj_positive_diff
        elif self.proj_type == 'euclid':
            self.objective_func = self._obj_l2_norm
        else:
            print(f"Php chiu {self.objective_func} khng ci t, chn 'qplus' hoc 'euclid'")

    def _obj_l2_norm(self, x, y):
        """Hm mc tiu khong cch Euclid chun (Dng cho C)"""
        return np.sqrt(np.sum((x - y)**2))
    
    def _obj_positive_diff(self, x, y):
        """Hm mc tiu c bit (Dng cho Qplus)"""
        v = np.maximum(x - y, 0) 
        return np.sum(v**2)

    def project(self, target_point):
        """
        init_point: im khi to thuc min cons
        target_point: im cn chiu 
        """
        init_point = np.random.rand(1, self.n).tolist()[0]
        
        res = minimize(
            self.objective_func,
            init_point,
            args=(target_point),
            jac="2-point",
            hess=BFGS(),
            constraints=self.cons,
            method="trust-constr",
            bounds=self.bounds,
            options={'disp': False}
        )
        
        optim_point = res.x
        
        if self.proj_type == 'qplus':
            return target_point - np.maximum(target_point - optim_point, 0)
        else:
            return optim_point
```

```python
def CQ_split_acceptance(f, jac_f, proj_C, proj_Qplus, x0,
                        gamma=1.0, max_iter=1000, tol=1e-6):
    print(f"Khi to: x0: {x0[0]}")
    x = proj_C(x0)
    print(f"Chiu ln C c: x: {x}")
    x_hist = [x.copy()]
    f_hist = [f(x).copy()]      # Lu gi tr f(x) thc t
    z_proj_hist = []  
    table = PrettyTable()
    table.field_names = ["k", "x_new", "y", "z_proj", "e_x", "e_f"]
    
    for k in tqdm(range(max_iter)):
        z = f(x)
        z_proj = proj_Qplus(z)          # P_{Q}(f(x))
        g = jac_f(x).T @ (z - z_proj)  # (x)
        x_new = proj_C(x - gamma * g)

        table.add_row([f"{k}", f"{x_new}", f"{z}", f"{z_proj}", f"{LA.norm(x_new - x)}", f"{LA.norm(z - z_proj)}"])
            
        if LA.norm(x - x_new) < tol:
            break
            
        x = x_new
        x_hist.append(x.copy())
        f_hist.append(f(x).copy())
        z_proj_hist.append(z_proj.copy())
    print(table)
    return x, np.array(x_hist), np.array(f_hist), np.array(z_proj_hist)
```

```python
c1_dx = grad(c1)
c2_dx = grad(c2)

q1_dx = grad(q1)
q2_dx = grad(q2)
q3_dx = grad(q3)
q4_dx = grad(q4)
```

```python
cons_C = (
    {
        'type': 'ineq',
        'fun' : lambda x: np.array([-c1(x)]),     
        'jac' : lambda x: np.array([-c1_dx(x)])   
    },
    
    {
        'type': 'ineq',
        'fun' : lambda x: np.array([-c2(x)]),     
        'jac' : lambda x: np.array([-c2_dx(x)])
    }
)

n = 2 #dim x
bounds_x = Bounds([-15,-15],[30, 30])
```

```python
cons_Q = (
    {
        'type': 'ineq',
        'fun' : lambda x: np.array([-q1(x)]),     
        'jac' : lambda x: np.array([-q1_dx(x)])   
    },
    
    {
        'type': 'ineq',
        'fun' : lambda x: np.array([-q2(x)]),     
        'jac' : lambda x: np.array([-q2_dx(x)])
    },
    
    {
        'type': 'ineq',
        'fun' : lambda x: np.array([-q3(x)]),     
        'jac' : lambda x: np.array([-q3_dx(x)])   
    },
    
    {
        'type': 'ineq',
        'fun' : lambda x: np.array([-q4(x)]),     
        'jac' : lambda x: np.array([-q4_dx(x)])
    }
)

bounds_f = None
```

```python
proj_C = Projection(cons=cons_C, bounds=bounds_x, n=n, proj_type='euclid')

proj_Qplus = Projection(cons=cons_Q, bounds=bounds_f, n=n, proj_type='qplus')

prob = Problem(
    f=[f1, f2],
    jac_f=[grad(f1), grad(f2)],
    C=[c1, c2],
    Q=[q1, q2, q3, q4],
    proj_C=proj_C.project,
    proj_Qplus=proj_Qplus.project
    
)
```

```python
x0 = np.random.rand(1,n).tolist()
```

```python
x_opt, x_hist, f_hist, z_proj_hist = CQ_split_acceptance(
                                f=prob.objective_func,
                                jac_f=prob.jacobian,
                                proj_C=prob.proj_C,
                                proj_Qplus=prob.proj_Qplus,
                                x0=x0,
                                gamma=0.01,
                                max_iter=50,
                                tol=1e-6
                            )
```

```python
def f1_vec(X, Y): return 4*X**2 + 4*Y**2
def f2_vec(X, Y): return (X - 5)**2 + (Y - 5)**2

x_range = np.linspace(-6, 12, 400)
y_range = np.linspace(-8, 12, 400)
X, Y = np.meshgrid(x_range, y_range)

val_c1 = (X - 5)**2 + Y**2 - 25
val_c2 = -(X - 8)**2 - (Y + 3)**2 + 7.7
mask_C = (val_c1 <= 0) & (val_c2 <= 0)

# Ly mu nn cho Plot 2
num_samples = 200000
x_rand = np.random.uniform(-6, 12, num_samples)
y_rand = np.random.uniform(-8, 12, num_samples)
c1_rand = (x_rand - 5)**2 + y_rand**2 - 25
c2_rand = -(x_rand - 8)**2 - (y_rand + 3)**2 + 7.7
mask_C_rand = (c1_rand <= 0) & (c2_rand <= 0)
x_valid = x_rand[mask_C_rand]
y_valid = y_rand[mask_C_rand]
f1_vals = f1_vec(x_valid, y_valid)
f2_vals = f2_vec(x_valid, y_valid)

# --- Bt u v ---
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

# ---------------- PLOT 1: KHNG GIAN BIN (x1, x2) ----------------
ax1.set_title("Qu o $x_k$ trn min C", fontsize=14, fontweight='bold')
ax1.contourf(X, Y, mask_C, levels=[0.5, 1.5], colors=['#A0C4FF'], alpha=0.6)
ax1.contour(X, Y, val_c1, levels=[0], colors='blue', linewidths=1)
ax1.contour(X, Y, val_c2, levels=[0], colors='red', linestyles='dashed', linewidths=1)

# >>> V QU O X <<<
ax1.plot(x_hist[:,0], x_hist[:,1], 'k.-', linewidth=1, markersize=3, label='Qu o $x_k$')
ax1.plot(x_hist[0,0], x_hist[0,1], 'g*', markersize=15, label='Start') # im u
ax1.plot(x_hist[-1,0], x_hist[-1,1], 'r*', markersize=15, label='End') # im cui

ax1.set_xlabel("$x_1$", fontsize=12)
ax1.set_ylabel("$x_2$", fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.5)
ax1.set_aspect('equal')
ax1.legend(loc='upper left')

# ---------------- PLOT 2: KHNG GIAN MC TIU (f1, f2) ----------------
ax2.set_title("Qu o $f(x_k)$ v $z_{proj}$ trn min Q", fontsize=14, fontweight='bold')

# Nn: nh ca C
ax2.scatter(f1_vals, f2_vals, s=5, color='#A0C4FF', alpha=0.4, label='nh ca min C')

# Nn: Hp Q
rect_Q = mpatches.Rectangle((20, 10), 130, 30, linewidth=2, edgecolor='red', facecolor='none', linestyle='--')
ax2.add_patch(mpatches.Rectangle((20, 10), 130, 30, color='#FFADAD', alpha=0.3))
ax2.add_patch(rect_Q)

# >>> V QU O f(x) (QUAN TRNG NHT) <<<
# y l gi tr mc tiu thc t di chuyn t ngoi vo trong hp
ax2.plot(f_hist[:,0], f_hist[:,1], 'k.-', linewidth=1, markersize=3, label='Qu o $f(x_k)$')

# >>> V QU O z_proj (Theo yu cu) <<<
# y l im chiu ln hp Q (s chy dc bin hp)
# V m hn hoc nh hn  khng ri
ax2.scatter(z_proj_hist[:,0], z_proj_hist[:,1], s=10, c='orange', marker='x', label='Mc tiu tm thi $z_{proj}$')

# im u v cui ca f(x)
ax2.plot(f_hist[0,0], f_hist[0,1], 'g*', markersize=15, label='Start $f(x_0)$')
ax2.plot(f_hist[-1,0], f_hist[-1,1], 'r*', markersize=15, label='End $f(x^*)$')

ax2.set_xlabel("$f_1(x)$", fontsize=12)
ax2.set_ylabel("$f_2(x)$", fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.5)
ax2.set_xlim(-10, 500) # Zoom out mt cht  thy im xut pht nu xa
ax2.set_ylim(-10, 150)
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
