# Problem

```python
def f1(x):
    return 4*x[0]**2 + 4*x[1]**2
def f2(x):
    return (x[0] - 5)**2 + (x[1] - 5)**2

#--------------- C --------------------#
def c1(x):
    return (x[0] - 5)**2 + x[1]**2 - 25
def c2(x):
    return -(x[0] - 8)**2 - (x[1] + 3)**2 + 7.7


#--------------- Q --------------------#
def q1(x):
    return -f1(x) + 20
def q2(x):
    return f1(x) - 150

def q3(x):
    return -f2(x) + 10
def q4(x):
    return f2(x) - 40

```

# Gradient approach

# Implement

```python
from scipy.optimize import minimize
from scipy.optimize import Bounds
from scipy.optimize import BFGS
from autograd import grad
import numpy as np
```

```python
class Problem():
    def __init__(self, f, jac_f, C, Q, proj_C, proj_Qplus):
        self.f = f
        self.jac_f = jac_f
        self.C = C
        self.Q = Q
        self.proj_C = proj_C
        self.proj_Qplus = proj_Qplus
        
class Projection():
    def __init__(self, cons, bounds, n):
        self.cons = cons
        self.bounds = bounds
        self.n = n
        
    def euclid(self, x, y):
        """The Euclid function"""
        return np.sqrt(np.sum((x-y)**2))
    
    def project(self, x_proj):
        """
        x_init: im khi to tm min
        x_proj: im cn chiu ln C
        """
        x_init = np.random.rand(1,n).tolist()
        res = minimize(
            self.euclid,
            x_init,
            args=(x_proj),
            jac="2-point",
            hess=BFGS(),
            constraints=self.cons,
            method="trust-constr",
            bounds=self.bounds,
            options={'disp': False}
        )
        return res.x
```

```python
c1_dx = grad(c1)
c2_dx = grad(c2)

q1_dx = grad(q1)
q2_dx = grad(q2)
q3_dx = grad(q3)
q4_dx = grad(q4)
```

```python
cons_C = (
    {
        'type': 'ineq',
        'fun' : lambda x: np.array([-c1(x)]),     
        'jac' : lambda x: np.array([-c1_dx(x)])   
    },
    
    {
        'type': 'ineq',
        'fun' : lambda x: np.array([-c2(x)]),     
        'jac' : lambda x: np.array([-c2_dx(x)])
    }
)

bounds_x = Bounds([-15,-15],[30, 30])
```

```python
cons_Q = (
    {
        'type': 'ineq',
        'fun' : lambda x: np.array([-q1(x)]),     
        'jac' : lambda x: np.array([-q1_dx(x)])   
    },
    
    {
        'type': 'ineq',
        'fun' : lambda x: np.array([-q2(x)]),     
        'jac' : lambda x: np.array([-q2_dx(x)])
    },
    
    {
        'type': 'ineq',
        'fun' : lambda x: np.array([-q3(x)]),     
        'jac' : lambda x: np.array([-q3_dx(x)])   
    },
    
    {
        'type': 'ineq',
        'fun' : lambda x: np.array([-q4(x)]),     
        'jac' : lambda x: np.array([-q4_dx(x)])
    }
)

bounds_f = None
```

```python
proj_C = Projection(
    cons = cons_C,
    bounds = bounds_x,
    n = n
)

proj_Qplus = Projection(
    cons = cons_Q,
    bounds = bounds_f,
    n = n
)

prob = Problem(
    f=[f1, f2],
    jac_f=[grad(f1), grad(f2)],
    C=[c1, c2],
    Q=[q1, q2, q3, q4],
    proj_C=proj_C,
    proj_Qplus=proj_Qplus
    
)
```

```python
def CQ_split_acceptance(f, jac_f, proj_C, proj_Qplus, x0,
                        gamma=1.0, max_iter=1000, tol=1e-6):
    x = proj_C.project(x0)
    for k in range(max_iter):
        z = f(x)
        z_proj = proj_Qplus.project(z)          # P_{Q}(f(x))
        g = jac_f(x).T @ (z - z_proj)  # (x)
        x_new = proj_C.project(x - gamma * g)
        if norm(x_new - x) < tol:
            break
        x = x_new
    return x
```

```python
n = 2
x0 = np.random.rand(1,n).tolist()
```

```python
CQ_split_acceptance(
    f=prob.f,
    jac_f=prob.jac_f,
    proj_C=prob.proj_C,
    proj_Qplus=prob.proj_Qplus,
    x0=x0
)
```
