# Gen

```python
num_region = 25 # = số nghiệm

start = np.pi/4
end = np.pi/2
```

* Random in partition

```python
rays = []
for j in range(num_region):
    random = np.random.uniform(start + j*(end-start)/num_region, start+ (j+1)*(end-start)/num_region)
    ray = np.array([np.cos(random),
                    np.sin(random)], dtype='float64')
    rays.append(ray)
```

* Sin cos

```python
rays = []
partitions = np.linspace(start, end, num_region)
for i in range(num_region):
    partition = partitions[i]
    ray = np.array([np.cos(partition),
                  np.sin(partition)], dtype='float32')
    rays.append(ray)
```

* linspace

```python
rays = []
start = 0
end = 1
partitions = np.linspace(start, end, num_region)
for partition in partitions:
    ray = np.array([1 - partition, partition])
    rays.append(ray)
rays
```

* angle

```python
import math
def calculate_angle_with_ox(vector):
    """Tính góc của vectơ với trục Ox, góc từ 0 đến pi/2"""
    x, y = vector
    angle = math.atan2(y, x)
    if angle < 0:
        angle += math.pi
    return angle

def find_max_min_angles(a, b):
    """Tìm góc lớn nhất và nhỏ nhất của các vectơ tạo bởi a và b"""
    vectors = [
        (a[0], b[1]),
        (b[0], a[1]),
    ]
    
    angles = [calculate_angle_with_ox(vec) for vec in vectors]
    
    min_angle = min(angles)
    max_angle = max(angles)
    
    return min_angle, max_angle
```

```python
def find_tangent_vectors(x, y, r):
    # Khoảng cách từ gốc tọa độ đến tâm đường tròn
    d = math.sqrt(x**2 + y**2)
    
    if d > r:
        # Tính góc alpha của đường từ gốc tọa độ đến tâm đường tròn
        alpha = math.atan2(y, x)
        
        # Tính góc beta giữa đường nối tâm và véctơ tiếp tuyến
        beta = math.asin(r / d)
        
        # Tính góc của hai tiếp tuyến
        angle1 = alpha + beta
        angle2 = alpha - beta
        
        # Tọa độ các điểm tiếp tuyến
        vector1 = (d * math.cos(angle1), d * math.sin(angle1))
        vector2 = (d * math.cos(angle2), d * math.sin(angle2))
        
        return [vector1, vector2]
    else:
        return []
    
    
vector1, vector2 = find_tangent_vectors(30, 30, 27.5)
min_x = min(vector1[0], vector2[0])
max_x = max(vector1[0], vector2[0])
min_y = min(vector1[1], vector2[1])
max_y = max(vector1[1], vector2[1])
```

```python
def circle_points(K, min_angle_=None, max_angle_=None):
    # generate evenly distributed preference vector
    ang0 = 1e-6 if min_angle_ is None else min_angle
    ang1 = np.pi / 2 - ang0 if max_angle_ is None else max_angle
    angles = np.linspace(ang0, ang1, K, endpoint=True)
    x = np.cos(angles)
    y = np.sin(angles)
    return np.c_[x, y]


min_angle = 0.1
max_angle = np.pi / 2 - 0.1

min_angle, max_angle = find_max_min_angles([min_x, min_y], [max_x, max_y])
print(min_angle, max_angle)
rays = circle_points(25, min_angle_=min_angle, max_angle_=max_angle)
rays
```

# Write

```python
from ruamel.yaml import YAML
from ruamel.yaml.comments import CommentedSeq
```

```python

# 3. Khởi tạo đối tượng YAML để giữ format và comment
yaml = YAML()
yaml.preserve_quotes = True
yaml.indent(mapping=2, sequence=4, offset=2) # Căn chỉnh lề cho đẹp

# 4. Đọc file config hiện tại
config_file = 'config.yaml'

try:
    with open(config_file, 'r', encoding='utf-8') as f:
        data = yaml.load(f)
except FileNotFoundError:
    print(f"Không tìm thấy file {config_file}, hãy tạo file mẫu trước.")
    exit()

# 5. Chuyển đổi dữ liệu Numpy sang định dạng YAML list chuẩn
#    Lưu ý: YAML không hiểu numpy.float, phải chuyển về float thường
yaml_rays = CommentedSeq()

for r in rays:
    # Tạo list con [x, y]
    # Dùng float() để convert từ numpy float64
    item = CommentedSeq([float(r[0]), float(r[1])])
    
    # Quan trọng: Thiết lập flow_style=True để nó ra dạng [x, y] nằm ngang
    item.fa.set_flow_style() 
    
    yaml_rays.append(item)

# 6. Gán dữ liệu mới vào cấu trúc config
if 'data' not in data:
    data['data'] = {}

data['data']['test_ray'] = yaml_rays

# 7. Ghi đè lại vào file config.yaml
with open(config_file, 'w', encoding='utf-8') as f:
    yaml.dump(data, f)

print(f"Đã cập nhật {len(rays)} tia mới vào file {config_file} thành công!")
```
