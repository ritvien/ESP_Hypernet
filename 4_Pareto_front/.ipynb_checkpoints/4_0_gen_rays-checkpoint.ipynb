# Prob

```python
def f(x):
    return np.array([
        4*x[0]**2 + 4*x[1]**2,
        (x[0] - 5)**2 + (x[1] - 5)**2
    ])


#--------------- C --------------------#
def c1(x):
    return -(x[0] - 5)**2 - x[1]**2 + 25
def c2(x):
    return -(x[0] - 8)**2 - (x[1] + 3)**2 + 17.7


#--------------- Q --------------------#
def q1(y):
    return 50**2 - (y[0] - 50)**2 - (y[1] - 50)**2

def q_plus(y):
    center = 50.0
    radius_sq = 50**2  
    
    dx = np.maximum(0, y[0] - center)
    dy = np.maximum(0, y[1] - center)
    return radius_sq - (dx**2 + dy**2)

# Hàm dùng cho Projection 
cons_C = (
    {
        'type': 'ineq', 'fun' : c1,     
    },
    {
        'type': 'ineq', 'fun' : c2,     
    }
)

dim_x = 2
bounds_x = Bounds([-15,-15],[30, 30])

cons_Q = (
    {
        'type': 'ineq', 'fun' : q1,     
    },    
)

cons_Qplus = (
    {'type': 'ineq', 'fun': q_plus},
)
dim_y = 2
```

```python
# Setup Projections
proj_C_handler = Projection(cons=cons_C, bounds=bounds_x, dim=dim_x, proj_type='euclid')
proj_Q_handler = Projection(cons=cons_Q, bounds=None, dim=dim_y, proj_type='qplus')

# Setup Problem
prob = Problem(
    f=[f], jac_f=[jacobian(f)], C=[], Q=[], 
    dim_x=dim_x, dim_y=dim_y,
    proj_C=proj_C_handler.project,
    proj_Qplus=proj_Q_handler.project
)
```

```python
cfg = load_config()
x_feasible, x_hist_p1, f_hist_p1, z_proj_hist = solve_CQ_feasible(
    f=prob.objective_func,
    jac_f=prob.jacobian,
    proj_C=prob.proj_C,
    proj_Qplus=prob.proj_Qplus,
    x0=cfg['initialization']['x_init'],
    gamma=cfg['phase1']['gamma'], 
    max_iter=cfg['phase1']['max_iter'],
    tol=cfg['phase1']['tol']
)
print(f"-> Điểm khả thi (Feasible Point): {x_feasible}")
```

```python
x_final, path_x = optim_Universal(
    prob=prob,
    x_feasible=x_feasible,  
    r=None,
    target_dim=1,
    mode='min',
    max_iter=cfg['phase2']['max_iter'],
    mu=cfg['phase2']['mu'],
    expo_alpha=cfg['phase2']['expo_alpha'],
    expo_lambda=cfg['phase2']['expo_lambda'],
    init_params=cfg['phase2']['init_params']
)
```

```python
mode_run = ["min", "max"]
dim_run = [0, 1]

limit_Q = []
for mode in mode_run:
    for dim in dim_run:
        x_final, _ = optim_Universal(
                prob=prob,
                x_feasible=x_feasible,  
                r=None,
                target_dim=dim,
                mode=mode,
                max_iter=cfg['phase2']['max_iter'],
                mu=cfg['phase2']['mu'],
                expo_alpha=cfg['phase2']['expo_alpha'],
                expo_lambda=cfg['phase2']['expo_lambda'],
                init_params=cfg['phase2']['init_params']
            )
        limit_Q.append(f(x_final)[dim])
```

```python
def find_max_min_angles(min_vals, max_vals):
    """
    Tìm góc quét của không gian mục tiêu dựa trên các giá trị cực biên.
    
    Args:
        min_vals: [min_x, min_y] (Giá trị nhỏ nhất tìm được của f1 và f2)
        max_vals: [max_x, max_y] (Giá trị lớn nhất tìm được của f1 và f2)
        
    Returns:
        (min_angle, max_angle): Góc nhỏ nhất và lớn nhất (radian) so với trục Ox
    """
    min_x, min_y = min_vals
    max_x, max_y = max_vals
    
    vectors = [
        (max_x, min_y),
        (min_x, max_y),
    ]
    print(vectors)
    
    angles = [math.atan2(vec[1], vec[0]) for vec in vectors]
    
    return min(angles), max(angles)
```

```python
def circle_points(K, min_angle_=None, max_angle_=None):
    # generate evenly distributed preference vector
    ang0 = 1e-6 if min_angle_ is None else min_angle
    ang1 = np.pi / 2 - ang0 if max_angle_ is None else max_angle
    angles = np.linspace(ang0, ang1, K, endpoint=True)
    x = np.cos(angles)
    y = np.sin(angles)
    return np.c_[x, y]
```

```python
min_vals = [min_x, min_y]
max_vals = [max_x, max_y]

min_angle, max_angle = find_max_min_angles(min_vals, max_vals)
# min_angle, max_angle = np.pi/4, np.pi/2
print(min_angle, max_angle)
rays = circle_points(25, min_angle_=min_angle, max_angle_=max_angle)
rays
```

```python
plt.figure(figsize=(8, 8)) # Kích thước hình vuông để không bị méo

# Vẽ đường tròn đơn vị (Unit Circle) làm nền để dễ hình dung
theta = np.linspace(0, np.pi/2, 100)
plt.plot(np.cos(theta), np.sin(theta), color='gray', linestyle='--', alpha=0.5, label='Unit Circle')

# Vẽ các Rays (Dùng Quiver để vẽ mũi tên vector)
# Cú pháp: quiver(x_gốc, y_gốc, u_hướng, v_hướng, ...)
origin_x = np.zeros(len(rays))
origin_y = np.zeros(len(rays))

plt.quiver(origin_x, origin_y,     # Điểm bắt đầu (0,0)
           rays[:, 0]*100, rays[:, 1]*100, # Hướng vector (x, y)
           angles='xy', scale_units='xy', scale=1, # Tỷ lệ chuẩn 1:1
           color='blue', width=0.005, label='Rays')

# Vẽ các điểm đích (để nhìn rõ sự phân bố)
plt.scatter(rays[:, 0], rays[:, 1], color='red', s=30, zorder=5)

# --- 4. Trang trí ---
plt.xlim(-0.1, 100) # Giới hạn trục X
plt.ylim(-0.1, 100) # Giới hạn trục Y
plt.axhline(0, color='black', linewidth=1) # Trục hoành
plt.axvline(0, color='black', linewidth=1) # Trục tung
plt.grid(True, linestyle=':', alpha=0.6)   # Lưới
plt.gca().set_aspect('equal')              # Quan trọng: Giữ tỷ lệ 1:1 để đường tròn không bị méo thành hình elip

plt.title(f'Visualization of {K} Rays on Unit Circle')
plt.xlabel('x (f1)')
plt.ylabel('y (f2)')
plt.legend(loc='upper right')

plt.show()
```

```python
from ruamel.yaml import YAML
from ruamel.yaml.comments import CommentedSeq
```

```python

# 3. Khởi tạo đối tượng YAML để giữ format và comment
yaml = YAML()
yaml.preserve_quotes = True
yaml.indent(mapping=2, sequence=4, offset=2) # Căn chỉnh lề cho đẹp

# 4. Đọc file config hiện tại
config_file = 'config.yaml'

try:
    with open(config_file, 'r', encoding='utf-8') as f:
        data = yaml.load(f)
except FileNotFoundError:
    print(f"Không tìm thấy file {config_file}, hãy tạo file mẫu trước.")
    exit()

# 5. Chuyển đổi dữ liệu Numpy sang định dạng YAML list chuẩn
#    Lưu ý: YAML không hiểu numpy.float, phải chuyển về float thường
yaml_rays = CommentedSeq()

for r in rays:
    # Tạo list con [x, y]
    # Dùng float() để convert từ numpy float64
    item = CommentedSeq([float(r[0]), float(r[1])])
    
    # Quan trọng: Thiết lập flow_style=True để nó ra dạng [x, y] nằm ngang
    item.fa.set_flow_style() 
    
    yaml_rays.append(item)

# 6. Gán dữ liệu mới vào cấu trúc config
if 'data' not in data:
    data['data'] = {}

data['data']['test_ray'] = yaml_rays

# 7. Ghi đè lại vào file config.yaml
with open(config_file, 'w', encoding='utf-8') as f:
    yaml.dump(data, f)

print(f"Đã cập nhật {len(rays)} tia mới vào file {config_file} thành công!")
```
